# 各国指数のパフォーマンス比較
このノートブックでは、yfinanceを使用して2025年1月以降の各国の主要株価指数を取得し、そのパフォーマンスを比較します。

```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# ティッカーシンボルに新しい指数を追加
indices = {
    'Nikkei 225': '^N225',
    'Dow Jones': '^DJI',
    'S&P 500': '^GSPC',
    'Nasdaq': '^IXIC',
    'Russell 2000': '^RUT',
    'SOX': '^SOX',
    # 'TOPIX': '^TOPX',
    # '東証グロース': '^TOKYO_GROWTH',
    # 'TOPIX500': '^TOPIX500',
    '台湾加権指数': '^TWII'
```

```python
# データ取得
data = {}
for name, ticker in indices.items():
    hist = yf.Ticker(ticker).history(start='2025-01-01')
    hist.index = hist.index.date  # タイムゾーンを削除
    data[name] = hist['Close']

# データフレームに統合
df = pd.DataFrame(data)
```

```python
import japanize_matplotlib
from datetime import datetime
# 正規化してパフォーマンスを比較
normalized_df = df.iloc[:, 1:] / df.iloc[0, 1:] * 100
normalized_df['Nikkei 225'] = df.iloc[:, 0] / df.iloc[2, 0] * 100
normalized_df.loc[[datetime(2025, 1, 2), datetime(2025, 1, 3)], 'Nikkei 225'] = 100

# 可視化
plt.figure(figsize=(14, 5))
for column in normalized_df.columns:
    plt.plot(normalized_df.index, normalized_df[column], label=column)

plt.title('各国指数のパフォーマンス比較 (2025年1月以降)')
plt.xlabel('日付')
plt.ylabel('パフォーマンス (%)')
plt.legend()
plt.grid()
plt.show()
```

# 米10年債利回りと主要株価指数の比較
このノートブックでは、yfinanceを使用して過去20年間の米10年債利回り、ダウジョーンズ、ナスダック、S&P 500のパフォーマンスを比較します。

```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# ティッカーシンボル
assets = {
    '10年債利回り': '^TNX',
    'Dow Jones': '^DJI',
    'Nasdaq': '^IXIC',
    'S&P 500': '^GSPC'
}

# データ取得
start_date = '2005-01-01'
data = {}
for name, ticker in assets.items():
    hist = yf.Ticker(ticker).history(start=start_date)
    hist.index = hist.index.date  # タイムゾーンを削除
    data[name] = hist['Close']

# データフレームに統合
df = pd.DataFrame(data)

# 正規化して比較
# df_normalized = df / df.iloc[0] * 100

# 可視化
fig, ax1 = plt.subplots(figsize=(14, 7))

# 第1軸: 株価指数
ax1.set_xlabel('日付')
ax1.set_ylabel('正規化パフォーマンス (%)', color='tab:blue')
for column in ['Dow Jones', 'Nasdaq', 'S&P 500']:
    ax1.plot(df.index, df[column], label=column)
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.legend(loc='upper left')

# 第2軸: 10年債利回り
ax2 = ax1.twinx()
ax2.set_ylabel('10年債利回り (%)', color='tab:red')
ax2.plot(df.index, df['10年債利回り'], label='10年債利回り', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')
ax2.legend(loc='upper right')

# タイトルとグリッド
plt.title('米10年債利回りと主要株価指数の比較 (過去20年)')
plt.grid()
plt.show()
```

# 日経平均225銘柄のパフォーマンス比較
このノートブックでは、yfinanceを使用して2025年3月以降の日経平均225銘柄の株価を取得し、そのパフォーマンスを比較します。

```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import requests
from bs4 import BeautifulSoup

# 日経平均225構成銘柄を取得する関数
def get_nikkei225_tickers():
    url = "https://indexes.nikkei.co.jp/nkave/index/component?idx=nk225"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # テーブルを取得
    table = soup.find_all('table')
    tickers = dict()
    for i in range(len(table)):
        rows = table[i].find_all('tr')[1:]  # ヘッダーをスキップ
        for row in rows:
            cols = row.find_all('td')
            tickers[cols[1].text.strip()] = cols[0].text.strip() + '.T'  # ティッカーシンボルを取得

    return tickers

# 日経平均225銘柄のティッカーリストと企業名（例として一部のみ記載）
n225_companies = get_nikkei225_tickers()

# データ取得
start_date = "2025-03-01"
data = {}
for name, ticker in n225_companies.items():
    print(f"Fetching data for {name} ({ticker})")
    hist = yf.Ticker(ticker).history(start=start_date)
    data[name] = hist['Close']

# データフレームに統合
df = pd.DataFrame(data)

# 正規化して比較
df_normalized = df / df.iloc[0] * 100

# グラフ作成
num_companies = len(n225_companies)
num_plots = (num_companies + 9) // 10  # 1グラフに最大10銘柄

for i in range(num_plots):
    plt.figure(figsize=(14, 7))
    companies_subset = list(n225_companies.keys())[i * 10:(i + 1) * 10]
    for company in companies_subset:
        plt.plot(df_normalized.index, df_normalized[company], label=company)
    plt.title(f'日経平均225銘柄の株価パフォーマンス比較 (グループ {i + 1})')
    plt.xlabel('日付')
    plt.ylabel('正規化パフォーマンス (%)')
    plt.yticks(ticks=np.arange(df_normalized.min().min() - 5, df_normalized.max().max() + 5, 5))  # メモリを2.5%刻みに設定
    plt.legend()
    plt.grid()
    plt.show()
```
